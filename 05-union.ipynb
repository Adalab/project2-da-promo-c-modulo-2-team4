# Unirlos

# txt+xml

```python
df_xml_txt = df_xml.merge(df_txt, on='index')
df_xml_txt.head(2)
```

# xml(genero) + txt +db(Q10)

```python
df_xml_txt_db = df_xml_txt.merge(df_db, on='index')
df_xml_txt_db.head(2)
```

# xml(genero) + txt +db(Q10) + kaggle ( el resto separadas)

```python
df_xml_txt_db_csv = df_xml_txt_db.merge(df_csv, on='index')
df_xml_txt_db_csv.head(2)
```

```python
#guardarlo en un archivo
df_xml_txt_db_csv.to_csv("data/05-data_union.csv")
```

```python
def union(txt,xml,df_db,df_csv):#db,kagle, txt. xml
    u_1= df_xml.merge(df_txt, left_on = "index", right_on = "index")#txt + xml
    u_2=u_1.merge(df_db, left_on = "index", right_on = "index")#db
    u_fin=u_2.merge(df_csv, left_on = "index", right_on = "index")#kaggle
    return u_fin
```

```python
df = union(df_db,df_csv,df_txt,df_xml)
```
