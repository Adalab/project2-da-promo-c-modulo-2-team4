## Funciones para abrir los archivos

Apertura

En este apartado generamos funciones que automaticen la apertura de los distintos archivos y generamos unos DataFrames con ellos.

```python
def apertura_csv (nombre_archivo):
    ruta = "../data-originales/" + nombre_archivo
    df_csv = pd.read_csv(ruta, index_col = 0)
    return df_csv
```

```python
def apertura_xml (nombre_archivo):
    ruta = "../data-originales/" + nombre_archivo
    df_xml = pd.read_xml(ruta)
    return df_xml
```

```python
def apertura_txt (nombre_archivo):
    ruta = "../data-originales/" + nombre_archivo
    df_txt = pd.read_csv(ruta, sep = ';')
    return df_txt
```

```python
def apertura_db (nombre_archivo):
    ruta = "../data-originales/" + nombre_archivo
    df_db = pd.read_csv(ruta)
    return df_db
```

```python
df_csv = apertura_csv("00-original-data_remaning_kaggle.csv")
df_csv.head(1)
```

```python
df_txt = apertura_txt("00-original-data_txt_clean.txt")
df_txt.head(1)
```

```python
df_db = apertura_db('00-original-exported_db_data.csv')
df_db.head(1)
```

```python
df_xml = apertura_xml('00-original-data_xml_clean.xml')
df_xml.head(1)
```

## Función para Guardar los archivos como CSV

Con esta función pretendemos facilitar el guardado de los archivos en un mismo formato para posteeriormente poder unirlos en uno solo y poder trabajar con toda la información necesaria.

```python
def guardar (df,nombre_guardado):
    ruta = "../data"
    os.chdir(ruta)
    return df.to_csv(nombre_guardado + ".csv")
```

```python
guardar(df_csv,"1.archivo_csv")
```

```python
guardar(df_txt,"2.archivo_txt")
```

```python
guardar(df_db,"3.archivo_db")
```

```python
guardar(df_xml,"4.archivo_xml")
```

# Unión automatizada de los archivos

Creamos una función que reciba los distintos archivos en el orden correcto para automatizar su unión mediante el método join(). Pretendemos así que sea más sencillo y rápido para el usuario. 

```python
def union(df_txt,df_xml,df_db,df_csv):#db,kagle, txt. xml
    df_txt.set_index("index", inplace =True)
    df_xml.set_index("index", inplace =True)
    df_db.set_index("index", inplace =True)
    df_csv.set_index("index", inplace =True)

    u_1= df_csv.join(df_db, on = "index")#txt + xml gender 
    u_2=u_1.join(df_txt, on = "index")#db Q10
    u_fin=u_2.join(df_xml, on = "index")#kaggle Q18
    return u_fin
```

```python
df = union(df_db,df_csv,df_txt,df_xml)
```

```python
#guardardo del archivo unido a través de la función guardar
guardar(df, "5.dataframe_unido") # La función decibe el dataframe y el nombre del archivo
```
